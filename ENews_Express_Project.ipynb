## 1. Do the users spend more time on the new landing page than the existing landing page?

### Perform Visual Analysis

```python
# lets plot a visual display of the time spent on the new page and the time spent on the existing page

plt.figure(figsize=(10,8))
sns.boxplot(data = landingpage_dataset, x = 'landing_page', y = 'time_spent_on_the_page')
plt.show()
```

### Step 1: Define the null and alternate hypotheses

Let $\mu_1, \mu_2$ represent the mean time spent by the users on both the new and existing landing page

We will test the null hypothesis

>$H_0:\mu_1=\mu_2$

against the alternate hypothesis

>$H_a:\mu_1>\mu_2$

### Step 2: Select Appropriate test

#### Two Independent Sample t-test

### Step 3: Decide the significance level

As given in the problem statement, significace level = 5% i.e,   =0.05 .

### Step 4: Collect and prepare data

## Lets calculate the standard deviation time spent by users on both the new and existing landing page by creating a subsetted data frame

```python
# Time spent on new landing page by users

spent_time_new_landingpage = landingpage_dataset[landingpage_dataset['landing_page'] == 'new']['time_spent_on_the_page']

# Time spent on existing landing page by users

spent_time_existing_landingpage = landingpage_dataset[landingpage_dataset['landing_page'] == 'old']['time_spent_on_the_page']
```

```python
# To print the standard deviation of the time spent for both new and existing landing page

print(
    'The standard deviation of the time spent on the new page is:', round(spent_time_new_landingpage.std(),2)
    )
print(
    'The standard deviation of the time spent on the existing page is:', round(spent_time_existing_landingpage.std(),2)
    )
```

###Insight

The population standard deviation for both groups are unequal

### Step 5: Calculate the p-value

```python
# To calculate the p-value using ttest_ind()

from scipy.stats import ttest_ind

test_stat, p_value = ttest_ind(spent_time_new_landingpage, spent_time_existing_landingpage, equal_var = False, alternative = 'greater')

print('The p_value of the hypothesis is', p_value)
```

### Step 6: Compare the p-value with $\alpha$

```python
# Lets print our conclusion based on p-value obtained above

if p_value < 0.05:
    print(
        f'Since the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'Since the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

### Step 7:  Draw inference

Since the p-value (~0.00014) is less than the level of significance, we reject the null hypothesis. Hence, we do have enough evidence to support claim that the users spent more time on the new landing page than the existing landing page

## 2. Is the conversion rate (the proportion of users who visit the landing page and get converted) for the new page greater than the conversion rate for the old page?

## Lets Perform Visual Analysis

```python
# Lets compare the conversion rate for the new page and the conversion rate for the old page using a bar chart

pd.crosstab(landingpage_dataset['converted'],landingpage_dataset['landing_page'],normalize='index').plot(kind="bar", figsize=(10,7),stacked=True)

plt.legend()
plt.show()
```

## Step 1: Lets define the null and alternate hypotheses

Let $\mu_1, \mu_2$ represent the mean conversion rate of the users on both the new and existing landing page

We will test the null hypothesis

>$H_0:\mu_1=\mu_2$

against the alternate hypothesis

>$H_a:\mu_1>\mu_2$

## Step 2: Select Appropriate test

#### Two proportions Z-test

## Step 3: Decide the significance level

As given in the problem statement, significace level = 5% i.e,   =0.05 .

## Step 4: Collect and prepare data

### Lets calculate the number of converted users in the treatment and control group

```python
# to calculate the number of converted users in the treatment group

number_converted_new = landingpage_dataset[landingpage_dataset['group'] == 'treatment']['converted'].value_counts()['yes']
number_converted_new
```

```python
# To calculate the number of converted users in the control group

number_converted_old = landingpage_dataset[landingpage_dataset['group'] == 'control']['converted'].value_counts()['yes']
number_converted_old
```

### Lets calculate the total number of users in both the treatment group and control group

```python
# total number of users in the treatment group

total_num_treatment = landingpage_dataset.group.value_counts()['treatment']

total_num_treatment
```

```python
# total number of users in the control group

total_num_control = landingpage_dataset.group.value_counts()['control']
total_num_control
```

```python
print('The numbers of users available in the treatment and control group are {0} and {1} respectively'.format(total_num_treatment, total_num_control))
```

## Step 5: Calculate the p-value

```python
# complete the code to import the required function
from statsmodels.stats.proportion import proportions_ztest

# write the code to calculate the p-value
test_stat, p_value = proportions_ztest([number_converted_new, number_converted_old] , [total_num_treatment, total_num_control], alternative = 'larger')

print(
    'The p_value for the hypothesis is', p_value
    )
```

## Step 6: Compare the p-value with  

```python
# To compare the p-value with the level of significance

if p_value < 0.05:
    print(f'Since the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'Since the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

## Step 7: Draw inference

Since the p_value (~0.008) is less than the level of significance, we reject the null hypothesis. Hence, we do have enough evidence to support the claim that, the conversion rate for the new landing page is greater than the conversion rate for the old landing page

## 3. Is the conversion and preferred language are independent or related?

## Step 1: Define the null and alternate hypotheses

### Let's write the null and alternative hypothesis


We will test the null hypothesis

>$H_0:$ The conversion and preferred language are independent

against the alternate hypothesis

>$H_a:$ The conversion and preferred language are related

## Step 2: Select Appropriate test

### Chi-square test of independence

## Step 3: Decide the significance level

As given in the problem statement, significace level = 5% i.e, =0.05 .

## Step 4: Collect and prepare data

```python
# lets create a contingency table showing the distribution of the two categorical variables

contingency_dataset = pd.crosstab(landingpage_dataset['converted'], landingpage_dataset['language_preferred'])

contingency_dataset
```

## Step 5: Calculate the p-value

```python
# import the required function
from scipy.stats import chi2_contingency
```

```python
# To calculate the p-value

chi2, p_value, dof, exp_freq = chi2_contingency(contingency_dataset)

print(
    'The p-value of the hypothesis is', p_value
    )
```

## Step 6: Compare the p-value with  

```python
# lets print the conclusion of our hypothesis based on p-value

if p_value < 0.05:
    print(f'Since the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'Since the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

## Step 7: Draw inference

Since the p_value (~0.213) is greater than the level of significance, we fail to reject the null hypothesis. Hence, we do not have enough evidence to support the claim that, the conversion status is related to the preferred language of the users

## 4. Is the time spent on the new page same for the different language users?

## Lets Perform Visual Analysis

### Lets create a new dataframe for the new landing page users

```python
# DataFrame for users on the new landing page

new_dataframe = landingpage_dataset[landingpage_dataset['landing_page'] == 'new']
```

```python
# Lets plot the time spent on the new page for different language users using a boxplot

plt.figure(figsize=(10,10))
sns.boxplot(data = new_dataframe, x = 'language_preferred', y = 'time_spent_on_the_page', showmeans = True)
plt.show()
```

```python
# Lets calculate the mean time spent on the new page for different language users

new_dataframe.groupby(['language_preferred'])['time_spent_on_the_page'].mean()
```

## Step 1: Define the null and alternate hypotheses

Let $\mu_1, \mu_2, \mu_3$ represent the mean time spent by the users on the new page with respect to their preferred language

We will test the null hypothesis

>$H_0: \mu_1 = \mu_2 = \mu_3$

against the alternate hypothesis

>$H_a: $ At least one of the preferred language users spent more time than other language users on the new page

## Step 2: Select Appropriate test

### One-way ANOVA Test

## Step 3: Decide the significance level

As given in the problem statement, significace level = 5% i.e, =0.05 .

## Step 4: Collect and prepare data

### Lets create a data frame for the time spent on the new page with respect to users preferred language

```python
# data frame of the time spent on the new page by English language users
english_time_spent = new_dataframe[new_dataframe['language_preferred']=="English"]['time_spent_on_the_page']


# data frame of the time spent on the new page by French language users
french_time_spent = new_dataframe[new_dataframe['language_preferred']=="French"]['time_spent_on_the_page']


# data frame of the time spent on the new page by Spanish language users
spanish_time_spent = new_dataframe[new_dataframe['language_preferred']=="Spanish"]['time_spent_on_the_page']
```

## Step 5: Calculate the p-value

```python
# lets import the ttest() function from scipy.stats
from scipy.stats import f_oneway
```

```python
test_stat, p_value = f_oneway(english_time_spent, french_time_spent, spanish_time_spent)

print(
    'The p_value for the hypothesis is', p_value
    )

```

## Step 6: Compare the p-value with  

```python
# Lets print the conclusion of our hypothesis based on p-value

if p_value < 0.05:
    print(f'Since the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'Since the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

## Step 7: Draw inference

Since the p_value (~0.43) is greater than the level of significance, we fail to reject the null hypothesis. Hence, we do not have enough evidence to support the claim that, users of one of the preferred languages spent more time on the new page than other preferred language users on the new page
